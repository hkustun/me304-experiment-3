```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
pd.options.mode.copy_on_write = True
```

TABLE MANIPULATION

```python
df = pd.read_excel("data3.xlsx")
df.columns
df1 = df[['Al(6mm)(N)', 'UP', 'DOWN', 'UP.1', 'DOWN.1']]
df2 = df[['Bronze(6mm)(N)', 'UP.2','DOWN.2', 'UP.3', 'DOWN.3']]
df3 = df[['Steel(3mm)(N)', 'UP.4', 'DOWN.4', 'UP.5','DOWN.5']]
df4 = df[['Steel(4.5mm)(N)', 'UP.6', 'DOWN.6', 'UP.7', 'DOWN.7']]
df5 = df[['Steel(6mm)(N)', 'UP.8', 'DOWN.8', 'UP.9', 'DOWN.9']]
df1
```

AVERAGE CALCULATIONS

```python
df1["average"] = (df1["DOWN"]+df1["DOWN.1"]+df1["UP"]+df1["UP.1"])/4
df2["average"] = (df2["DOWN.2"]+df2["DOWN.3"]+df2["UP.2"]+df2["UP.3"])/4
df3["average"] = (df3["DOWN.4"]+df3["DOWN.5"]+df3["UP.4"]+df3["UP.5"])/4
df4["average"] = (df4["DOWN.7"]+df4["DOWN.6"]+df4["UP.6"]+df4["UP.7"])/4
df5["average"] = (df5["DOWN.8"]+df5["DOWN.9"]+df5["UP.8"]+df5["UP.9"])/4
df1.at[4,"average"] = (df1.at[4,"UP"] + df1.at[4,"UP.1"]) / 2
df2.at[4,"average"] = (df2.at[4,"UP.2"] + df2.at[4,"UP.3"]) / 2
df3.at[4,"average"] = (df3.at[4,"UP.4"] + df3.at[4,"UP.5"]) / 2
df4.at[4,"average"] = (df4.at[4,"UP.6"] + df4.at[4,"UP.7"]) / 2
df5.at[4,"average"] = (df5.at[4,"UP.8"] + df5.at[4,"UP.9"]) / 2
df1
```

LEAST SQUARE LINE PLOTS FOR EVERY BEAM

```python
input_data = df1["Al(6mm)(N)"].values
measurement_data = df1["average"].values
#find slope and intercept of the line
slope,intercept = np.polyfit(input_data,measurement_data, 1)

# Create the least squares line
line = slope * input_data + intercept

# Plot the data points and the least squares line
plt.plot(input_data, line, color='black', label='Least squares line')
plt.scatter(input_data, df1["UP"], label='UP', color="red")
plt.scatter(input_data, df1["UP.1"], label='UP2',color="blue")
plt.scatter(input_data, df1["DOWN"], label='DOWN',color="red",marker="*")
plt.scatter(input_data, df1["DOWN.1"], label='DOWN2',color="blue",marker="*")

# Add labels and legend
plt.xlabel('Load(N)')
plt.ylabel('δmax(mm)')
plt.title('Least squares line for Al(6mm)')
plt.legend()
equation_text = f'y = {slope:.4f} * x + ({intercept:.4f})'
plt.text(0.6, 0.8, equation_text, transform=plt.gca().transAxes, fontsize=10, color='black', ha='center')

# Show the plot
plt.show()

df1["bestfit"] = df1["Al(6mm)(N)"] * slope + intercept
```

ACCURACY CALCULATIONS

```python
output_span_1 = df1.at[4,"bestfit"] - df1.at[0,"bestfit"]
output_span_2 = df2.at[4,"bestfit"] - df2.at[0,"bestfit"]
output_span_3 = df3.at[4,"bestfit"] - df3.at[0,"bestfit"]
output_span_4 = df4.at[4,"bestfit"] - df4.at[0,"bestfit"]
output_span_5 = df5.at[4,"bestfit"] - df5.at[0,"bestfit"]

df1.at[0,"max accuracy errors%(+ -)"] = max(df1["Max deviation from LSL"])/output_span_1*100
df1.at[1,"max accuracy errors%(+ -)"] = min(df1["Min deviation from LSL"])/output_span_1*100

df2.at[0,"max accuracy errors%(+ -)"] = max(df2["Max deviation from LSL"])/output_span_2*100
df2.at[1,"max accuracy errors%(+ -)"] = min(df2["Min deviation from LSL"])/output_span_2*100

df3.at[0,"max accuracy errors%(+ -)"] = max(df3["Max deviation from LSL"])/output_span_3*100
df3.at[1,"max accuracy errors%(+ -)"] = min(df3["Min deviation from LSL"])/output_span_3*100

df4.at[0,"max accuracy errors%(+ -)"] = max(df4["Max deviation from LSL"])/output_span_4*100
df4.at[1,"max accuracy errors%(+ -)"] = min(df4["Min deviation from LSL"])/output_span_4*100

df5.at[0,"max accuracy errors%(+ -)"] = max(df5["Max deviation from LSL"])/output_span_5*100
df5.at[1,"max accuracy errors%(+ -)"] = min(df5["Min deviation from LSL"])/output_span_5*100
```

HYSTERESIS AND REPEATABILITY CALCULATIONS FOR EACH BEAM

```python
#REPEATABILITY
for index,row in df1.iterrows(): 
    df1.at[index,"repeatability"] = max((max(row[2],row[4]) - min(row[2],row[4])),(max(row[1],row[3]) - min(row[1],row[3])))
for index,row in df2.iterrows(): 
    df2.at[index,"repeatability"] = max((max(row[2],row[4]) - min(row[2],row[4])),(max(row[1],row[3]) - min(row[1],row[3])))
for index,row in df3.iterrows():  
    df3.at[index,"repeatability"] = max((max(row[2],row[4]) - min(row[2],row[4])),(max(row[1],row[3]) - min(row[1],row[3])))
for index,row in df4.iterrows():  
    df4.at[index,"repeatability"] = max((max(row[2],row[4]) - min(row[2],row[4])),(max(row[1],row[3]) - min(row[1],row[3])))
for index,row in df5.iterrows():  
    df5.at[index,"repeatability"] = max((max(row[2],row[4]) - min(row[2],row[4])),(max(row[1],row[3]) - min(row[1],row[3])))
#HYSTERESIS
for index,row in df1.iterrows():
    df1.at[index,"hysteresis"] = max(abs(max(row[2],row[4]) - min(row[1],row[3])),abs(max(row[1],row[3])-min(row[2],row[4])))
for index,row in df2.iterrows():
    df2.at[index,"hysteresis"] = max(abs(max(row[2],row[4]) - min(row[1],row[3])),abs(max(row[1],row[3])-min(row[2],row[4])))
for index,row in df3.iterrows():
    df3.at[index,"hysteresis"] = max(abs(max(row[2],row[4]) - min(row[1],row[3])),abs(max(row[1],row[3])-min(row[2],row[4])))
for index,row in df4.iterrows():
    df4.at[index,"hysteresis"] = max(abs(max(row[2],row[4]) - min(row[1],row[3])),abs(max(row[1],row[3])-min(row[2],row[4])))
for index,row in df5.iterrows():
    df5.at[index,"hysteresis"] = max(abs(max(row[2],row[4]) - min(row[1],row[3])),abs(max(row[1],row[3])-min(row[2],row[4])))

df1.at[4,"repeatability"] = abs(df1.at[4,"UP"] - df1.at[4,"UP.1"])
df2.at[4,"repeatability"] = abs(df2.at[4,"UP.2"] - df2.at[4,"UP.3"])
df3.at[4,"repeatability"] = abs(df3.at[4,"UP.4"] - df3.at[4,"UP.5"])
df4.at[4,"repeatability"] = abs(df4.at[4,"UP.6"] - df4.at[4,"UP.7"])
df5.at[4,"repeatability"] = abs(df5.at[4,"UP.8"] - df5.at[4,"UP.9"])
```

```python
df1.at[0,"hysteresis error%"] = max(df1["hysteresis"])/output_span_1 * 100
df1.at[0,"repeatability error%"] = max(df1["repeatability"])/output_span_1 * 100

df2.at[0,"hysteresis error%"] = max(df2["hysteresis"])/output_span_2 * 100
df2.at[0,"repeatability error%"] = max(df2["repeatability"])/output_span_2 * 100

df3.at[0,"hysteresis error%"] = max(df3["hysteresis"])/output_span_3 * 100
df3.at[0,"repeatability error%"] = max(df3["repeatability"])/output_span_3 * 100

df4.at[0,"hysteresis error%"] = max(df4["hysteresis"])/output_span_4 * 100
df4.at[0,"repeatability error%"] = max(df4["repeatability"])/output_span_4 * 100

df5.at[0,"hysteresis error%"] = max(df5["hysteresis"])/output_span_5 * 100
df5.at[0,"repeatability error%"] = max(df5["repeatability"])/output_span_5 * 100
```

COMPLIANCE

```python
data = {
    'beam': ['steel3mm','steel4.5mm','steel6mm','brass6mm','aluminum6mm'],
    'E(GPa)': [210,210,210,105,76],
    'length(m)': [0.675,0.675,0.675,0.675,0.675],
    'width(m)': [0.019,0.019,0.019,0.019,0.019],
    'height(m)': [0.003,0.0045,0.006,0.006,0.006]
}

dfbeam = pd.DataFrame(data)
dfbeam["I(m^4)"] = dfbeam["width(m)"] * dfbeam["height(m)"]**3 / 12
#since compliance is defined as deflection/force, we can easily say that comp = slope / 1000 (we divide it by 1000 because deflection is in mms) of bestfit. 
dfbeam.at[0,"compliance(s^2/kg)"] = slope3/1000
dfbeam.at[1,"compliance(s^2/kg)"] = slope4/1000
dfbeam.at[2,"compliance(s^2/kg)"] = slope5/1000
dfbeam.at[3,"compliance(s^2/kg)"] = slope2/1000
dfbeam.at[4,"compliance(s^2/kg)"] = slope/1000
#using comp, length and I, we can estimate young's modulus.
dfbeam["E(GPa)(experimental)"] = dfbeam["length(m)"]**3 / (48*dfbeam['I(m^4)']*dfbeam["compliance(s^2/kg)"]) / 10**9
dfbeam["E error%"] = abs(dfbeam["E(GPa)"] - dfbeam["E(GPa)(experimental)"]) / dfbeam["E(GPa)"] * 100
dfbeam
```
